```
import pandas as pd
import csv
import numpy as np
```

```
tsv_read_main = pd.read_csv(path1, sep='\t')
tsv_read_target = pd.read_csv(path2, sep='\t', keep_default_na=False, na_values=['_'])
```

```
tsv_read_target
```

```
tsv_read_main=tsv_read_main.drop(columns=['gene_id'])
```

```
tsv_read_main
```

```
geneid_col_drop=tsv_read_main.columns
```

```
for i in geneid_col_drop:           #to keep only columns from 01-09 tumor cells
    str=i[13:14]
    j=0
    if(int(str)!=j):
        tsv_read_main=tsv_read_main.drop(columns=i)
set(tsv_read_main.columns)
```

```
col_name=[]   #convert to list
col_name=list(tsv_read_main.columns)
```

```
TSS_col=[]              #to get code for cancer in list like 'OR'
for col in col_name:
    TSS_col.append(col[5:7])
set(TSS_col)    
```

```
new_df = pd.DataFrame(list(zip(col_name, TSS_col)), 
               columns =['Image_id', 'TSS_code'])
```

```
new_df
```

```
tsv_read_target=tsv_read_target.drop(columns=['Source Site','BCR'])
tsv_read_target
```

```
cancer_type=[]                  #FIND cancer values for patients
length1=len(new_df)
length2= len(tsv_read_target)

i=0
for i in range(length1):
  j=0
  while j<length2:
    if (new_df['TSS_code'][i]==tsv_read_target['TSS Code'][j]):
      cancer_type.append(tsv_read_target['Study Name'][j])
      #print(i,cancer_type[i])
      break
    else:
      j=j+1
```

```
z=np.unique(cancer_type)
z
```

```
df2 = pd.DataFrame(cancer_type,columns=['cancer_type']) 
#df2.to_csv('cancer_type.csv', encoding='utf-8')
#output_final = [new_df, df2]
result = new_df.join(df2)
#result.to_csv('output.csv', encoding='utf-8')
result.head(20)
```

```
cancer_type                    
i=0
elm_count=[]
while i<len(z):
  elm_count.append(cancer_type.count(z[i]))
  print('The count of element', z[i], elm_count[i])
  i=i+1
```

```
column_array=tsv_read_main.columns
```

```
for lt in column_array:
  tsv_read_main[lt]=np.log2(tsv_read_main[lt]+1) 
```

```
tsv_read_main
```

```
tsv_read_drop=tsv_read_main.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)
tsv_read_drop.reset_index(drop=True, inplace=True)
print(tsv_read_drop.shape)
tsv_read_drop
```

```
variance_cal=tsv_read_drop.var(axis=1)
```

```
var_list=list(variance_cal)
len_var=len(var_list)
var_df=pd.DataFrame(var_list,columns=['variance'])
```

```
len(var_list)
```

```
frames = [tsv_read_drop, var_df]
df_var_new = tsv_read_drop.join(var_df)
df_var_new.head(30)
```

```
test_var_thre=df_var_new
indexNames = test_var_thre[ test_var_thre['variance'] >= 1.19]
indexNames
```

```
indexNames.reset_index(drop=True, inplace=True)
indexNames[0:50]
```
